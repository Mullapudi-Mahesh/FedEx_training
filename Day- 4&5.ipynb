List

```python
nums = [1, 2, 2, 3, 4, 4, 4, 5]

freq = {}
for n in nums:
    freq[n] = freq.get(n, 0) + 1

sorted_freq = dict(sorted(freq.items(), key=lambda x: x[1], reverse=True))

print("Number Frequency:")
for k, v in sorted_freq.items():
    print(f"{k} → {v} times")
```

Dictionary

```python
students = {
    "Likith": 85,
    "Arun": 92,
    "Sita": 78,
    "Ravi": 92
}

ranked = sorted(students.items(), key=lambda x: x[1], reverse=True)

print(" Student Rankings")
for rank, (name, marks) in enumerate(ranked, start=1):
    print(f"Rank {rank}: {name} ({marks})")
```

Banking System (Encapsulation + Validation)

```python
class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.__balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return " Invalid deposit amount"
        self.__balance += amount
        return f" Deposited ₹{amount}"

    def withdraw(self, amount):
        if amount > self.__balance:
            return " Insufficient balance"
        self.__balance -= amount
        return f" Withdrawn ₹{amount}"

    def get_balance(self):
        return f" Balance: ₹{self.__balance}"

acc = BankAccount("Likith", 5000)
print(acc.deposit(3000))
print(acc.withdraw(2000))
print(acc.get_balance())
```

Polymorphism(Notification System)

```python
class Notification:
    def send(self):
        pass

class Email(Notification):
    def send(self):
        return " Email sent"

class SMS(Notification):
    def send(self):
        return " SMS sent"

class WhatsApp(Notification):
    def send(self):
        return " WhatsApp message sent"

channels = [Email(), SMS(), WhatsApp()]

for c in channels:
    print(c.send())
```

Error Handling

Input Validation + Retry Mechanism

```python
while True:
    try:
        age = int(input("Enter age: "))
        if age <= 0:
            raise ValueError("Age must be positive")
        print(f" Age accepted: {age}")
        break
    except ValueError as e:
        print(" Error:", e)
```

Custom Exception (Enterprise Style)


```python
class InsufficientFundsError(Exception):
    pass

balance = 5000
amount = int(input("Enter withdrawal amount: "))

try:
    if amount > balance:
        raise InsufficientFundsError("Not enough balance")
    print(" Transaction successful")
except InsufficientFundsError as e:
    print("", e)
```

File Handling

Log User Actions

```python
from datetime import datetime

def log_action(user, action):
    with open("app.log", "a") as file:
        file.write(f"{datetime.now()} | {user} | {action}\n")

log_action("Likith", "Login")
log_action("Likith", "Viewed Dashboard")

print(" Logs written successfully")
```

Read & Analyze Log File

```python
with open("app.log", "r") as file:
    logs = file.readlines()

print(" Log Summary")
print("Total Actions:", len(logs))

for log in logs:
    print(log.strip())
```
